```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("/content/synthetic_fb_ads_undergarments.csv")
df.head()
```

```python
print("Average CTR:", df["ctr"].mean())
print("Average ROAS:", df["roas"].mean())
```

```python
daily = df.groupby("date").agg({"ctr":"mean","roas":"mean"}).reset_index()
daily
```

```python
daily.plot(x="date", y="ctr", title="CTR Trend", figsize=(6,3))
plt.show()
```

```python
daily.plot(x="date", y="roas", title="ROAS Trend", figsize=(6,3))
plt.show()
```
